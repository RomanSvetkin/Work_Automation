```python
import pandas as pd
import numpy as np
from glob import glob

all_files = glob('U2C_*.csv')

li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    li.append(df)

U2C = pd.concat(li, axis=0, ignore_index=True)
```

```python
U2C = U2C.drop(['Company Name for Emails', 'Email Status','Email Confidence', 'Contact Owner', 
                     'Work Direct Phone','Home Phone', 'Mobile Phone', 'Corporate Phone','Other Phone','Stage',
                     'Lists', 'Last Contacted', 'Account Owner', '# Employees', 'Industry', 
                     'Keywords', 'Person Linkedin Url', 'Website', 'Company Linkedin Url',
                     'Facebook Url', 'Twitter Url', 'City','State', 'Country', 'Company Address',
                     'Company City','Company State', 'Company Country', 'Company Phone','SEO Description', 'Technologies',
                     'Annual Revenue', 'Email Sent', 'Email Open', 'Email Bounced', 'Replied', 'Demoed',
                     'Number of Retail Locations', 'Apollo Contact Id', 'Apollo Account Id'], axis = 1)
```

```python
U2C
```

```python
U2C = U2C.rename({'First Phone': 'Phone Number', 'Title': 'Position'}, axis = 1)
```

```python
U2C
```

```python
U2C['Name'] = U2C["First Name"].astype(str) + ' ' + U2C["Last Name"].astype(str)
```

```python
U2C = U2C.drop(['First Name', 'Last Name'], axis = 1)
```

```python
U2C.isnull().sum()
```

```python
U2C = U2C.dropna()
```

```python
U2C.isnull().sum()
```

```python
U2C.head(20)
```

```python
U2C['Phone Number'] = U2C['Phone Number'].astype('string').str.replace("'", "").str.replace("%","").str.replace("-"," ").str.replace("(", "").str.replace(")","").str.replace(" ","").str.strip()
```

```python
U2C
```

```python
U2C['Temp'] = U2C['Phone Number'].str.len().gt(9)
```

```python
U2C.head(20)
```

```python
U2C = U2C[U2C["Temp"] == True]
```

```python
U2C.head(20)
```

```python
U2C = U2C.drop(['Temp'], axis = 1)
```

```python
U2C.head(20)
```

```python
U2C_msp_1 = U2C[U2C["Position"].astype('string').str.contains('founder', regex=False, case = False)]
U2C_msp_2 = U2C[U2C["Position"].astype('string').str.contains('CEO', regex=False, case = False)]
U2C_msp_3 = U2C[U2C["Position"].astype('string').str.contains('director', regex=False, case = False)]
U2C_msp_4 = U2C[U2C["Position"].astype('string').str.contains('Chief*', regex=True, case = False)]
U2C_msp_5 = U2C[U2C["Position"].astype('string').str.contains('head*', regex=True, case = False)]
U2C_msp = pd.concat([U2C_msp_1, U2C_msp_2, U2C_msp_3, U2C_msp_4, U2C_msp_5], ignore_index=True)
```

```python
U2C_msp
```

```python
U2C_msp = U2C_msp.drop_duplicates(subset=['Company'], ignore_index=True)
U2C_msp = U2C_msp.drop_duplicates(subset=['Name'], ignore_index=True)
```

```python
U2C_msp
```

```python
U2C_sm_1 = U2C[U2C["Position"].astype('string').str.contains('marketing', regex=False, case = False)]
U2C_sm_2 = U2C[U2C["Position"].astype('string').str.contains('ecommerce', regex=False, case = False)]
U2C_sm = pd.concat([U2C_sm_1, U2C_sm_2], ignore_index=True)
```

```python
U2C_sm
```

```python
U2C_sm = U2C_sm.drop_duplicates(subset=['Company'], ignore_index=True)
U2C_sm = U2C_sm.drop_duplicates(subset=['Name'], ignore_index=True)
```
